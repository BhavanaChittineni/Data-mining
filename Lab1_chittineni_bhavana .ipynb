```python
#We like to impute missing values in this column with the help of other values in this data frame. First, group the data frame by the gym_frequency column and calculate each group's mean age.
#Round the mean values to the closest integer and cast them to integers.

each_group_mean_age = df.groupby('gym_frequency')['age'].mean().round().astype(int) 
print(each_group_mean_age)

#groupby('gym_frequency')['age'] runs like the unique values of gym_frequency have different groups and every group is having age column as specified 
#and then calculates mean value of age column for every each age group. and rounded the value and covert data type 

#Iterate through the rows of the DataFrame df. For each row, check if the age value is NaN and the gym_frequency matches specific conditions ('1 or 2 days a week', '3 to 5 days of a week', 'everyday' or 'never'). If the conditions are met, replace np.nan with the corresponding rounded mean age value computed before. Now enter the mean of column age with three digits after the decimal point.
def check_conditions(j):
    conditions = [
        (j['gym_frequency'] == '1 or 2 days a week'),
        (j['gym_frequency'] == '3 to 5 days of a week'),
        (j['gym_frequency'] == 'everyday'),
        (j['gym_frequency'] == 'never')
    ]
    return any(conditions)
for i, j in df.iterrows():
    if np.isnan(j['age']) and check_conditions(j):
        df.at[i, 'age'] = each_group_mean_age[j['gym_frequency']]
print(df['age'].mean().round(3))
```

```python
#8th question
#1st part 
p = df['children'].value_counts(normalize=True)
print(p) 
#here the normalization makes the process easy first df[children].value_counts() makes the unique value count and excluding missing values
#when normalization sets as true it gives the probability for each unique value by taking the count of unique value divide by total value count

```

```python
#1st part
non_missing_children = df['children'].dropna()
value_counts = non_missing_children.value_counts()
probabilities = value_counts / non_missing_children.shape[0]
print(probabilities)
```

```python
np.random.seed(0)
values = np.random.choice([0, 1, 2, 3, 4, 5], 
                          size=df['children'].isnull().sum(), 
                          p=[p[0], p[1], p[2], p[3], p[4], p[5]])
print(values)
missing_indices = df[df['children'].isnull()].index
value_index = 0
for index in missing_indices:
    df.at[index, 'children'] = values[value_index]
    value_index += 1
children_freq = df['children'].value_counts().sort_index()
print(children_freq)
```

```python
#16th question

# creating new column decades and chanfing age into decade 
df['decade'] = df['age'] // 10 * 10

#  by using groupby function we find out the median avg steps per day for each decade
decade_medians = df.groupby('decade')['avg_steps_per_day'].median()

# if avgstepsper day value is null then replacing it with decade medians value
df.loc[df['avg_steps_per_day'].isna(), 'avg_steps_per_day'] = df['decade'].map(decade_medians)

# last step is to print the meanavg steps per day value
mean_avg_steps_per_day = df['avg_steps_per_day'].mean()
print(f"Mean of avg_steps_per_day column: {mean_avg_steps_per_day:.2f}")
```
