```python
#read csv file 
import pandas as pd
df = pd.read_csv("updated_insurance.csv")
df.info()
```

```python
#1st method here directly assigning numerical values considering 1 as low,2 as affordable, 3 as high
#creating function charges_categorize and passing charges column
#for this method data type changes to object
#2nd method using for loop and assigning values
#for 2nd method datatype not changing
for i in df.index:
    if df.loc[i,'charges'] <= 4500:
        df.loc[i,'charges'] = 1 #low
    elif df.loc[i,'charges'] > 10000:
        df.loc[i,'charges'] = 3 #high
    else :
        df.loc[i,'charges'] = 2 #affordable

df['charges'].value_counts()




```

```python
X = df.drop('charges', axis=1)
y = df['charges']
X.info()
y.info()
```

```python
#3rd question :
#split the data into train and test with test size = 0.2 and random state is 0
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)


```

```python
#24 question
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Create a new DecisionTreeClassifier with max_depth=1 and random_state=0
clf = DecisionTreeClassifier(max_depth=1, random_state=0)

# Fit the classifier to the training data
clf.fit(X_train, y_train)

# Predict the labels for the training data
y_pred_train = clf.predict(X_train)

# Calculate the accuracy score for the training data
train_score = accuracy_score(y_train, y_pred_train)

# Predict the labels for the test data
y_pred_test = clf.predict(X_test)

# Calculate the accuracy score for the test data
test_score = accuracy_score(y_test, y_pred_test)

# Calculate the score gap between train and test scores
score_gap = abs(train_score - test_score)

# Print the score gap with two decimal places
print("Score gap: {:.2f}".format(score_gap))
```

```python
#25th question:
# Setting the highest value of score gap
score_gap = float('inf')
best_depth = 0

for depth in range(1,11):
    clf2 = DecisionTreeClassifier(max_depth=depth,random_state=0)
    clf2.fit(X_train, y_train)
    train_score = clf2.score(X_train, y_train)
    test_score = clf2.score(X_test, y_test)
    gap = np.abs(train_score - test_score)
    if train_score > 0.8 and test_score > 0.8 and gap < score_gap :
        score_gap = gap
        best_depth = depth
    
print("Max_depth:", best_depth)
```

```python
#29th question
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score

# Given X_train, X_test, y_train, and y_test
# Define the combinations of hyperparameters
hyperparameters = [
    {'max_depth': 2, 'max_samples': 0.5, 'max_features': 0.2},
    {'max_depth': 1, 'max_samples': 0.5, 'max_features': 0.2},
    {'max_depth': 1, 'max_samples': 0.5, 'max_features': 0.5},
    {'max_depth': 1, 'max_samples': 0.2, 'max_features': 0.2},
    {'max_depth': 2, 'max_samples': 0.5, 'max_features': 0.5},
    {'max_depth': 2, 'max_samples': 0.2, 'max_features': 0.5},
    {'max_depth': 1, 'max_samples': 0.2, 'max_features': 0.5},
    {'max_depth': 2, 'max_samples': 0.2, 'max_features': 0.2}
]

# Calculate train and test scores for each hyperparameter combination
for params in hyperparameters:
    rfc = RandomForestClassifier(**params, random_state=0)
    rfc.fit(X_train, y_train)

    y_train_pred = rfc.predict(X_train)
    train_macro_precision = precision_score(y_train, y_train_pred, average='macro')

    y_test_pred = rfc.predict(X_test)
    test_macro_precision = precision_score(y_test, y_test_pred, average='macro')

    print(f"Hyperparameters: max_depth={params['max_depth']}, max_samples={params['max_samples']}, max_features={params['max_features']}")
    print(f"Train macro-precision score: {train_macro_precision:.2f}")
    print(f"Test macro-precision score: {test_macro_precision:.2f}\n")
```
